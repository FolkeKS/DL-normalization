```python
data_dir = "data/processed/nemonemo_bnd_perten1_samples_standardize/"
```

```python
f = open(data_dir+"norms_std_mean.txt")
lines = f.readlines()
assert len(lines) == 2, f"len {len(lines)}"
std = float(lines[0])
mean = float(lines[1])
f.close()
```

```python
X_train = transforms.CenterCrop([310, 380])(torch.from_numpy(np.load(data_dir+"train/X/NORSSH_BND_PERTEN84.npz")['arr_0']).numpy()
Y_train = transforms.CenterCrop([310, 380])(torch.from_numpy(np.load(data_dir+"train/Y/NORSSH_BND_PERTEN84_norm_coeffs.npz")['arr_0']).numpy()
X_valid = transforms.CenterCrop([310, 380])(torch.from_numpy(np.load(data_dir+"valid/X/NORSSH_BND_PERTEN199.npz")['arr_0']).numpy()
Y_valid = transforms.CenterCrop([310, 380])(torch.from_numpy(np.load(data_dir+"valid/Y/NORSSH_BND_PERTEN199_norm_coeffs.npz")['arr_0']).numpy()
distance_map = transforms.CenterCrop([310, 380])(torch.from_numpy(np.load("data/sign_distance_map.npy")).numpy()
```

```python
def compute_eps(X,Y, model_path,distance_map=None): 

    if distance_map != None:
        X_map = np.empty((4,310,380))
        distance_map = transforms.CenterCrop([310, 380])(distance_map)
        X = transforms.CenterCrop([310, 380])(X)
        X = torch.cat((X,torch.unsqueeze(distance_map, 0)),0)
        cnn_map.CNN.load_from_checkpoint(model_path)  
        
    else :
        cnn.CNN.load_from_checkpoint(model_path)   
    
    
    Y = transforms.CenterCrop([290, 360])(torch.from_numpy(Y)).numpy()
    X = transforms.CenterCrop([310, 380])(torch.from_numpy(X[np.newaxis,:]).float())
    mask = np.where(Y==0,True,False)
    Y = Y*std + mean

    Y_pred = model.forward(X).detach()*std + mean
    Y_pred =  transforms.CenterCrop([290, 360])(Y_pred).numpy()[0,0,:,:]

    Y2 = np.power(Y,2)
    eps = (np.power(Y_pred,2) - Y2)/Y2
    eps = np.ma.masked_array(eps, mask)

    print("mean: ",np.mean(eps[np.nonzero(eps)]))
    print("max: ",np.max(eps))
    print("min: ",np.min(eps))
    
    return eps
```

```python
model_path = "results/wandb/cnn/1_1_adam/checkpoints/epoch=45222-val_loss=0.00020.ckpt"
eps = compute_eps(X_train,Y_train,model_path)
fig,axs = plt.subplots(figsize=(30, 10),ncols=1)
c = axs.imshow(eps,origin="lower",vmin=-0.05,vmax=0.05,cmap="RdYlBu")
cbar_ax = fig.add_axes([0.266, 0.05, 0.4673, 0.02])
plt.colorbar(c,cax=cbar_ax,orientation='horizontal')
plt.show()
```
